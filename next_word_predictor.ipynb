```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
# Sample corpus
text = """This is a simple next word prediction model using RNN. 
         It predicts the next word based on previous words. 
         Deep learning models like LSTMs are effective for this."""
```

```python

# Tokenize the text
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1
```

```python
# Create input sequences
input_sequences = []
for sentence in text.split('.'):
    words = sentence.strip().split()
    for i in range(1, len(words)):
        n_gram_sequence = words[:i+1]
        seq = tokenizer.texts_to_sequences([' '.join(n_gram_sequence)])[0]
        input_sequences.append(seq)
```

```python
# Pad sequences
max_seq_length = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_seq_length, padding='pre')
```

```python
import tensorflow as tf
# Split data into inputs and labels
X, y = input_sequences[:, :-1], input_sequences[:, -1]
y = tf.keras.utils.to_categorical(y, num_classes=total_words)
```

```python
# Build RNN Model
model = Sequential([
    Embedding(total_words, 10, input_length=max_seq_length-1),
    LSTM(100, return_sequences=True),
    LSTM(100),
    Dense(100, activation='relu'),
    Dense(total_words, activation='softmax')
])
```

```python
# Compile and Train Model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=100, verbose=1)
```

```python
# Next word prediction function
def predict_next_word(seed_text, num_words=1):
    for _ in range(num_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_seq_length-1, padding='pre')
        predicted_probs = model.predict(token_list, verbose=0)
        predicted_word_index = np.argmax(predicted_probs)
        for word, index in tokenizer.word_index.items():
            if index == predicted_word_index:
                seed_text += " " + word
                break
    return seed_text
```

```python
# Test Prediction
print(predict_next_word("It predicts the next"))
```
